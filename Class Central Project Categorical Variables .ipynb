```python
import numpy as np
import random
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
%matplotlib inline
```

```python
subjects=pd.read_csv('sub and classname.csv')
```

```python
# combine into one function input: a column in subjects, outputs a transforms another column
def courses_remove(subject_name_raw):
    subject_name= subject_name_raw.replace(' Courses', '')
    return subject_name
```

```python
#test=np.ones(len(subjects['subject_name']))
test=[None]*len(subjects['subject_name'])

for i, item in enumerate(subjects['subject_name']):
    temp= courses_remove(item)
    test[i]=str(temp)

subjects['Subjects']=test
```

```python
subjects
```

```python
sns.set(style='darkgrid')
ax = sns.countplot(y='Subjects', data=subjects, palette="Set3")
```

```python
subjects['Subjects'].describe()
```

```python
Inst= pd.read_csv('all_subjects_institution.csv')
```

```python
Inst
```

```python
# combine into one function input: a column in subjects, outputs a transforms another column
def inst_extract(inst_name_raw):
    inst_name=inst_name_raw[98:inst_name_raw.find('"institutionId"')].strip(' ').strip('"",\n')
    return inst_name
```

```python
#test=np.ones(len(subjects['subject_name']))
test=[None]*len(Inst['course_institution'])

for i, item in enumerate(Inst['course_institution']):
    temp= inst_extract(item)
    test[i]=str(temp)

Inst['Institution']=test
```

```python
Inst
```

```python
sns.countplot(y="Institution", data=Inst, palette="Set3",
              order=Inst.Institution.value_counts().iloc[1:15].index)
```

```python
fig, ax = plt.subplots()
color_palette_list = ['#009ACD', '#ADD8E6', '#63D1F4', '#0EBFE9',   
                      '#C1F0F6', '#0099CC']
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=14
labels = ['Courses With Certificate', 'Courses Without Certificate']

percentages = [10525/len(Inst['course_institution']), 1-(10525/len(Inst['course_institution']))]
explode=(0.1,0)
ax.pie(percentages, explode=explode, labels=labels,  
       colors=color_palette_list[0:2], autopct='%1.0f%%', 
       shadow=False, startangle=0,   
       pctdistance=1.2,labeldistance=1.4)
ax.axis('equal')
#ax.set_title("Class Central Cert")
ax.legend(frameon=False, bbox_to_anchor=(1.5,0.8))
```
